```python
%sql
CREATE OR REPLACE TABLE workspace.default.contracts_clean
USING DELTA
AS
SELECT
    CAST(contract_transaction_unique_key AS STRING) AS contract_transaction_unique_key,
    CAST(contract_award_unique_key AS STRING) AS contract_award_unique_key,
    CAST(award_id_piid AS STRING) AS award_id_piid,
    CAST(modification_number AS STRING) AS modification_number,
    CAST(transaction_number AS STRING) AS transaction_number,
    CAST(parent_award_agency_id AS STRING) AS parent_award_agency_id,
    CAST(parent_award_agency_name AS STRING) AS parent_award_agency_name,
    CAST(parent_award_id_piid AS STRING) AS parent_award_id_piid,
    CAST(parent_award_modification_number AS STRING) AS parent_award_modification_number,
    CAST(federal_action_obligation AS DECIMAL(18,2)) AS federal_action_obligation,
    CAST(total_dollars_obligated AS DECIMAL(18,2)) AS total_dollars_obligated,
    CAST(action_date AS DATE) AS action_date,
    CAST(action_date_fiscal_year AS INT) AS action_date_fiscal_year,
    CAST(period_of_performance_start_date AS DATE) AS period_of_performance_start_date,
    CAST(period_of_performance_current_end_date AS DATE) AS period_of_performance_current_end_date,
    CAST(period_of_performance_potential_end_date AS DATE) AS period_of_performance_potential_end_date
FROM workspace.default.raw_contracts
WHERE contract_transaction_unique_key IS NOT NULL
  AND contract_award_unique_key IS NOT NULL
  AND federal_action_obligation IS NOT NULL;
```
