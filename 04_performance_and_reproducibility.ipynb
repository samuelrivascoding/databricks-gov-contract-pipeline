```python
%sql
OPTIMIZE workspace.default.fact_contracts
ZORDER BY (parent_award_agency_id, action_date);
```

```python
%sql
CREATE OR REPLACE TABLE workspace.default.agency_obligation_summary
USING DELTA
AS
SELECT
    parent_award_agency_id,
    parent_award_agency_name,
    COUNT(*) AS num_transactions,
    SUM(federal_action_obligation) AS total_obligation,
    AVG(federal_action_obligation) AS avg_obligation,
    MIN(federal_action_obligation) AS min_obligation,
    MAX(federal_action_obligation) AS max_obligation
FROM workspace.default.fact_contracts
GROUP BY parent_award_agency_id, parent_award_agency_name;
```

```python
%sql
SELECT *
FROM workspace.default.agency_obligation_summary
WHERE total_obligation > 1000000
LIMIT 100;
```

```python
%sql
DESCRIBE DETAIL workspace.default.agency_obligation_summary;
```
